```python
run_id = 483
stat = RunStats.objects.get(pk=run_id)

periods = stat.periods.all()

dfs = {}

for p in periods:
    print(p.title)
    wts = Topic.objects.filter(period=p,run_id=stat.parent_run_id).annotate(
        mscore = Max('topicdtopic__score'),
        sscore = Sum('topicdtopic__score')
    ).order_by('sscore')
    
    dfs[p.title] = pd.DataFrame.from_dict(list(wts.values(
        'mscore',
        'sscore',
        'score',
        'title'
    )))
```
