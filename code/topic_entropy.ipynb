```python
import pandas as pd
import matplotlib.pyplot as plt

run_id = 1103
p = 50
```

```python
topic_info = pd.read_csv(f"../tables/topic_info_{run_id}_{p}.csv")
topic_info.head()
```

```python
var = "wg_entropy"

pdf = topic_info.sort_values(var,ascending=False).reset_index(drop=True)

pdf1 = pdf[:50]
pdf2 = pdf[50:]

fig, axs = plt.subplots(2,figsize=(12,18))

ax = axs[0]

pdf1[var].plot(ax=ax)
ax.set_xticks(pdf1.index)
ax.set_xticklabels(pdf1['topic__title'],rotation=45,ha="right")

ax = axs[1]

pdf2[var].plot(ax=ax)
ax.set_xticks(pdf2.index)
ax.set_xticklabels(pdf2['topic__title'],rotation=45,ha="right")

fig.tight_layout()

plt.savefig(f"../plots/topic_wg_entropy.png")
plt.show()
```
